# Power Analysis for AI writing study

### First of all

The outcomes of this study fall into three categories:
1. **behavioural process measures** during writing (eye fixation and revision activity),
2. **cognitive outcomes** reflecting internalization of the written argument (surface and structural recall)
3. **subjective authorship perceptions** (perceived ownership and willingness to disclose authorship)

Behavioural measures serve as **process evidence** of how participants engage with the writing task, whereas recall and authorship measures capture **cognitive and subjective consequences** of these interactions.

### One-way ANOVA or mixed-effects

The key dependent variable of this study is **"delayed structural recall"** (i.e. how many essay arguments participants remember 1 month after the second essay task).

For that single enpdoint:
- Independent variable (manipulation): condition (3 groups)
- Dependent variable (effect) : delayed structural recall

which we can treat as **one-way between-subject comparison** (e.g. ANOVA)


Meanwhile, the process evidence such as metacognitive behaviours (i.e. fixation time and revision amount),  immediate cognitive outcomes (surface and structural recall immediately after the task) and subjective authorship perception (perceived ownership and willingness to disclose authorship) are measured twice per person (session 1 and session 2). This means for these dependent variables (effect), we have

- Between-subject: condition
- Within-subject: session

as the independent variables (manipulation), which makes it **mixed-effects**. How groups change differently from session 1-2 is not as relevant, since this most likely shows learning effects with the tool/experiment set up. The main effect (conditional differences averaged across sessions) would serve as evidence to the core theoretical outcome of this study (i.e. delayed structural recall), but these are secondary outcomes.


### Why even run 2 essay writing sessions in the first place?

Kosmyna et al's "Your Brain on ChatGPT" suggested that repeated reliance on AI writing assistance may reduce cognitive engagement with the writing task. To account for this, we run the essay writing task twice to allow for the potential offloading effects to emerge. While Kosmyna et al used the multiple session set up to explore the "cumulative cognitive offloading effects of AI usage over time", we build on their finding and use the multiple session setup as the **"stabilization of cognitive offloading behaviour"**.

### Calculating the effect size for a one-way ANOVA with 3 groups

Cohen’s f is the effect size used for ANOVA, which corresponds to how separated the group means are relative to the within-group noise.

| Cohen’s f | Interpretation | Variance explained (η²) |
| :--- | :---: | ---: |
| 0.10 | small | ~1% |
| 0.25 | medium | ~6% |
| 0.40 | large | ~14% |

So if the power analysis says f = 0.41, it means the condition explains about **14–15% of the variance** in the outcome.

```python
from statsmodels.stats.power import FTestAnovaPower

analysis = FTestAnovaPower()

alpha = 0.05
power = 0.80
groups = 3

N = 60   # total sample size

f = analysis.solve_power(
    effect_size=None,
    nobs=N,
    alpha=alpha,
    power=power,
    k_groups=groups
)

print(f)
```

| N | Effect size (f) |
| :--- | :---: |
| 30 | f ≈ 0.6 (very large effects) |
| 45 | f ≈ 0.48 (still large effects) |
| 60 | f ≈ 0.41 (large effects) |
| 150 | f ≈ 0.25 (medium effects) |

### Aiming for a N=60 sample size

The power of this study is evaluated for the **primary endpoint (delayed structural recall after the second essay)**, analyzed as a 3-group between-subjects comparison.

A one-way between-subjects comparison across three conditions (α=0.05, power=0.80), indicates that N=60 provides 80% power to detect large interface effects (Cohen’s f≈0.41; η²≈0.14). If recruitment falls below this target, the study will be interpreted as a **proof-of-concept evaluation** primarily sensitive to larger effects (if N=45 then f ≈ 0.48, if N=30 then f ≈ 0.60).
